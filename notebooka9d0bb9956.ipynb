```python
# Data and Model: Simple Linear Regression
# Features: x1, x2
# Target: y = 3*x1 + 5*x2 + 2
import numpy as np

# Generate sample data
x1 = np.linspace(0, 10, 100)  # Feature 1
x2 = np.linspace(10, 20, 100)  # Feature 2
y = 3 * x1 + 5 * x2 + 2  # Target variable

# Mimicking Partial Dependence Plots (PDP)
# PDP calculates the average prediction as one feature varies, holding others constant.
pdp_x1 = [np.mean(3 * x + 5 * x2 + 2) for x in x1]
pdp_x2 = [np.mean(3 * x1 + 5 * x + 2) for x in x2]

# Mimicking SHAP values
# SHAP calculates contributions of individual features to the prediction.
# For simplicity, we will calculate the average contribution for each feature.
mean_x1_contrib = np.mean(3 * x1)  # Coefficient * feature value
mean_x2_contrib = np.mean(5 * x2)  # Coefficient * feature value

# Visualization (Optional)
import matplotlib.pyplot as plt

# Plot PDP for x1
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x1, pdp_x1, label="PDP for x1")
plt.xlabel("x1")
plt.ylabel("Average Prediction")
plt.title("Partial Dependence Plot for x1")
plt.legend()

# Plot PDP for x2
plt.subplot(1, 2, 2)
plt.plot(x2, pdp_x2, label="PDP for x2", color="orange")
plt.xlabel("x2")
plt.ylabel("Average Prediction")
plt.title("Partial Dependence Plot for x2")
plt.legend()

plt.tight_layout()
plt.show()

# Print "SHAP-like" values
print(f"Mean Contribution of x1: {mean_x1_contrib}")
print(f"Mean Contribution of x2: {mean_x2_contrib}")
```
